```python
pip install -U sentence-transformers
pip install tensorflow==2.3.0
!pip3 install --upgrade tensorflow-gpu

from sentence_transformers import SentenceTransformer
model = SentenceTransformer('bert-base-nli-mean-tokens')
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.', 
    'The quick brown fox jumps over the lazy dog.']
sentence_embeddings = model.encode(sentences)

print("Sentence embeddings:")
print(sentence_embeddings)

# https://huggingface.co/sentence-transformers/bert-base-nli-mean-tokens 
```

## Sentence_BERT Model <a class="anchor" id="sbert"></a>


```python
import pandas as pd
df_1 = pd.read_csv("data_v4.csv")
```


* News Headlines

```python
from sentence_transformers import SentenceTransformer
model_name = SentenceTransformer('bert-base-nli-mean-tokens')
sentences = df_1['News Headlines']
sentence_embeddings = model_name.encode(sentences)
print("Sentence embeddings:")
print(sentence_embeddings)
```

```python
df_textembeddings = pd.DataFrame(data=sentence_embeddings)
df_textembeddings
```

```python
import numpy as np
df_textembeddings.index = np.arange(1, len(df_textembeddings) + 1)
df_textembeddings
```

```python
df_textembeddings.to_csv('df_textembeddings.csv')
```

* News Content

```python
from sentence_transformers import SentenceTransformer
model_name = SentenceTransformer('bert-base-nli-mean-tokens')
sentences2 = df_1['News content']
sentence_embeddings2 = model_name.encode(sentences2)

df_textembeddings2 = pd.DataFrame(data=sentence_embeddings2)
df_textembeddings2
```

```python
import numpy as np
df_textembeddings2.index = np.arange(1, len(df_textembeddings2) + 1)
df_textembeddings2
```

```python
df_textembeddings2.to_csv('df_textembeddings2.csv')
```
